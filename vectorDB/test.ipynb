```python
import pandas as pd 
from datasets import load_dataset, concatenate_datasets
import numpy as np
import os
from unidecode import unidecode
```

```python
df = pd.read_csv("./data/vie_wiki_dataset.csv")
```

```python
keywords = [
    "IP", "máy tính", "kháng sinh", "vi khuẩn", "cốt lõi", 
    "công nghiệp", "Chế tài", "vi phạm", "quy định", "nguồn gốc", "điều trị", 
    "trầm cảm", "thần học", "lạm phát", "tổng cung", "Ngoại tệ", "dinh dưỡng", 
    "tim mạch", "ngôi chùa", "tiếng việt", "giao thông", "thương mại", "quốc tế", 
    "phân trang", "bộ nhớ", "thanh ghi", "co dãn", "hành chính", "phân quyền", 
    "tập quyền", "phân cấp", "tản quyền", "nhiễm sắc thể"
]

# Chuẩn hóa từ khóa (bỏ dấu và chuyển về lowercase)
keywords_normalized = [unidecode(kw).lower() for kw in keywords]

# Hàm kiểm tra xem text có chứa từ khóa không
def contains_keyword(text):
    if pd.isna(text):
        return False
    text_normalized = unidecode(str(text)).lower()
    return any(kw in text_normalized for kw in keywords_normalized)

# Lọc dataframe
df_filtered = df[df['text'].apply(contains_keyword)]

# Chỉ giữ lại các cột cần thiết
df_result = df_filtered[['id', 'title', 'text']]

print(f"Số dòng ban đầu: {len(df)}")
print(f"Số dòng sau khi lọc: {len(df_result)}")
print("\nMẫu dữ liệu:")
print(df_result.head())
```

```python
df1 = pd.read_csv("./data/wiki.csv")
df_result = df_result[~df_result['id'].isin(df1['id'])]
len(df_result)
```

```python
df2 = pd.concat([df1, df_result], ignore_index=True)
df2.to_csv('./data/wiki.csv', index=False)
```

```python
dfs = np.array_split(df_result, 3)
dfs[0].to_csv('./data/data1.csv', index=False)
dfs[1].to_csv('./data/data2.csv', index=False)
dfs[2].to_csv('./data/data3.csv', index=False)
```
