```python
import pandas as pd
df = pd.read_csv("./data/vie_wiki_dataset.csv")
df.head(3)
```

```python
len(df['title'].unique())
```

```python
# Danh sách từ khóa cần tìm
keywords = ["văn", "toán", "hóa", "lý", "lí", "triết", "học", "địa", 
            "kinh", "tế", "chính", "trị", "luật", "lịch", "sử"]

# Tạo pattern regex để tìm bất kỳ từ nào trong danh sách
# Sử dụng word boundary \b để tìm từ đầy đủ
pattern = '|'.join(keywords)

# Lọc dataframe
df_filtered = df[df['title'].str.contains(pattern, case=False, na=False, regex=True)]

# Loại bỏ duplicate theo cột 'id'
df_result = df_filtered.drop_duplicates(subset=['id'], keep='first')

print(f"Số lượng rows sau khi lọc: {len(df_filtered)}")
print(f"Số lượng rows sau khi loại duplicate: {len(df_result)}")
```

```python
df_result = df_result[df_result['id'] > 0]

print(f"Số lượng rows sau khi lọc id > 0: {len(df_result)}")
```

```python
# Tính số lượng rows cho mỗi phần
total_rows = len(df_result)
chunk_size = total_rows // 3

# Chia dataframe thành 3 phần
df_part1 = df_result.iloc[0:chunk_size]
df_part2 = df_result.iloc[chunk_size:chunk_size*2]
df_part3 = df_result.iloc[chunk_size*2:]

# Xuất ra 3 file CSV
df_part1.to_csv('data1.csv')
df_part2.to_csv('data2.csv')
df_part3.to_csv('data3.csv')
```
